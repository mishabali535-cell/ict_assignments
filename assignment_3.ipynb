Write a Python function to sum all the numbers in a list.
Sample List : [18, 12, 30,10, 7]
Expected Output : 77

```python
def sum_list_numbers(numbers):
    total = 0
    for x in numbers:
        total += x
    return total

sample_list = [18, 12, 30, 10, 7]
print(f"Sample List: {sample_list}")
print(f"Expected Output: {sum_list_numbers(sample_list)}")
```

3. Python program to insert a number to any position in a list (Get the number
to be inserted and position from the user)

```python
my_list = [10, 20, 40, 50]
print(f"Original List: {my_list}")
number_to_insert = int(input("Enter the number you want to insert: "))
position = int(input(f"Enter the position (0-{len(my_list)}) to insert the number: "))
my_list.insert(position, number_to_insert)
print(f"Updated List: {my_list}")
```

4. Write a program to rename a key city to a location in the following dictionary.
