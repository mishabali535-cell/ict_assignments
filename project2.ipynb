Personal Expense Tracker Develop a Personal Expense Tracker to monitor and analyze spending behavior. Users should be able to log expenses with date, category, and amount. The system should generate monthly summaries, category-wise spending reports, budget breach alerts, and identify high-spending patterns. Analytical insights like highest expense day and dominant spending category must be computed using loops and conditionals.

```python
expenses = []

BUDGET = 5000

while True:
    print("\n--- Personal Expense Tracker ---")
    print("1. Add Expense")
    print("2. Show Total Expense")
    print("3. Category-wise Expense")
    print("4. Highest Expense Day")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        date = input("Enter date (DD-MM): ")
        category = input("Enter category: ")
        amount = int(input("Enter amount: "))

        expenses.append([date, category, amount])
        print("Expense Added!")

    elif choice == "2":
        total = 0
        for e in expenses:
            total = total + e[2]

        print("Total Expense:", total)

        if total > BUDGET:
            print("⚠ Budget Exceeded!")

    elif choice == "3":
        categories = {}

        for e in expenses:
            if e[1] in categories:
                categories[e[1]] = categories[e[1]] + e[2]
            else:
                categories[e[1]] = e[2]

        print("Category-wise Expense:")
        for c in categories:
            print(c, ":", categories[c])

    elif choice == "4":
        day_total = {}

        for e in expenses:
            if e[0] in day_total:
                day_total[e[0]] = day_total[e[0]] + e[2]
            else:
                day_total[e[0]] = e[2]

        max_day = ""
        max_amount = 0

        for d in day_total:
            if day_total[d] > max_amount:
                max_amount = day_total[d]
                max_day = d

        if max_day != "":
            print("Highest Expense Day:", max_day, "-", max_amount)
        else:
            print("No expenses found")

    elif choice == "5":
        print("Thank you!")
        break

    else:
        print("Invalid choice")
```

Develop a Log File Analyzer for simulated system logs.
The program should parse text-based log entries, categorize events, count error types,
analyze user activity, and filter logs based on time or severity.
String splitting, pattern matching, and dictionary-based aggregation must be used. The
application should present analytical summaries via a menu-driven interface.

```python
records = [
    "INFO Alice Login",
    "ERROR Bob Failed",
    "INFO Alice Logout",
    "WARNING Alice Expire",
    "ERROR Bob Failed",
    "INFO Charlie Login"
]

level_count = {}
user_count = {}
error_count = {}

# Process records
for r in records:
    words = r.split(" ")

    level = words[0]
    user = words[1]
    action = words[2]

    # Count levels
    if level in level_count:
        level_count[level] = level_count[level] + 1
    else:
        level_count[level] = 1

    # Count users
    if user in user_count:
        user_count[user] = user_count[user] + 1
    else:
        user_count[user] = 1

    # Count errors
    if level == "ERROR":
        if action in error_count:
            error_count[action] = error_count[action] + 1
        else:
            error_count[action] = 1

# Menu
while True:
    print("\n--- Menu ---")
    print("1. Show level count")
    print("2. Show error count")
    print("3. Show user count")
    print("4. Filter by level")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        print("\nLevel Count")
        for k in level_count:
            print(k, ":", level_count[k])

    elif choice == "2":
        print("\nError Count")
        for k in error_count:
            print(k, ":", error_count[k])

    elif choice == "3":
        print("\nUser Count")
        for k in user_count:
            print(k, ":", user_count[k])

    elif choice == "4":
        lvl = input("Enter level (INFO/WARNING/ERROR): ")
        print("\nRecords:")
        for r in records:
            if r.startswith(lvl):
                print(r)

    elif choice == "5":
        print("Program finished")
        break

    else:
        print("Wrong choice")
```
